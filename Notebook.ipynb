```python
import os
from google.colab import userdata
from PyPDF2 import PdfReader
from langchain_core.documents import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Pinecone as PineconeSearch
from langchain.chains import RetrievalQA
from pinecone import Pinecone, ServerlessSpec
```

```python
def readText(pdfFiles):
  text=""
  for pdfFile in pdfFiles:
    reader=PdfReader(pdfFile)
    for page in reader.pages:
      text = text + page.extract_text()
  return text
```

```python
text=readText(['/content/TIMETABLE_IS 2020_2021 Semester V.pdf'])
text
```

```python
textChuncks=RecursiveCharacterTextSplitter(chunk_size=200,
                                           chunk_overlap=100,
                                           length_function=len
                                           ).split_text(text)

textChuncks[1]
```

```python
huggingFaceEmbeddings=HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L12-v2")
```

```python
os.environ['PINECORN_API_KEY']=userdata.get('PINECORN_API_KEY')
```

```python
pinecone=Pinecone(api_key=os.environ['PINECORN_API_KEY'])
```

```python
dimensions=384

pinecone.create_index(
    name="pdf-rag",
    dimension=dimensions,
    metric="cosine",
    spec=ServerlessSpec(
        cloud="aws",
        region="us-east-1"
    )
)
```

```python
pinecone.list_indexes()
```

```python
index=pinecone.Index("pdf-rag")

index.describe_index_stats()
```

```python
vectoreStore=PineconeSearch.from_texts(
    texts=textChuncks,
    embedding=huggingFaceEmbeddings,
    index_name="pdf-rag"
)
```
